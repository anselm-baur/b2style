```python
def is_2d_list(lst):
    # Check if the list is not empty and the first element is a list
    if isinstance(lst, list) and len(lst) > 0 and isinstance(lst[0], list):
        return True
    return False
    
def visualize_color_list(colors, color_sets=None, alphas=[1, 0.7, 0.55, 0.35]):
    import matplotlib.patches as patches
    import matplotlib.pyplot as plt

    if is_2d_list(alphas) and len(alphas) != len(colors):
        raise RuntimeError("list of alphas needs to be the same langth as list of colors!")

    color_palette = {}
    len_colors  =  len(colors)*len(alphas)
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(6, len_colors * 0.6))

    # Add color patches to the plot
    j = 0
    for i, color in enumerate(colors):
        # Create a rectangle for each color
        if color_sets is not None:
            color_set = color_sets[i]
        else:
            color_set = f"Color Set {i}"
        print(color_set)
        color_palette[color_set] = []
        if is_2d_list(alphas) and len(alphas) == len(colors):
            _alphas = alphas[i]
        else:
            _alphas = alphas
        for alpha in _alphas:
            nu_color = b2fig.colors.trans(color, alpha)
            rect = patches.Rectangle((0, len_colors - j - 1), 1, 1, facecolor=nu_color)
            ax.add_patch(rect)
            print(f"{color} alpha:  {alpha} -> {nu_color} ({_alphas})")
            color_palette[color_set].append(nu_color) #f" ({int(color[1:2], 16):2} {int(color[3:5], 16):2} {int(color[5:7], 16):2})"+ "\n"
            # Add a label next to each rectangle
            ax.text(1.1, len_colors - j - 0.5, f'Color {j+1}: {nu_color}', ha='left', va='center', fontsize=10)
            j += 1

    # Set the limits and remove ticks
    ax.set_xlim(0, 2)
    ax.set_ylim(0, len_colors)
    ax.set_xticks([])
    ax.set_yticks([])

    # Add a title
    plt.title('Visualization of Color List', pad=20)
    return color_palette
```

```python
def create_linear_cm(colors, name="my_color_map"):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.colors import LinearSegmentedColormap
    
    custom_cmap = LinearSegmentedColormap.from_list(name, colors)

    # Create some data to visualize the colormap
    data = np.outer(np.linspace(0, 1, 100), np.ones(100))
    
    # Plot the data using the custom colormap
    plt.imshow(data, aspect='auto', cmap=custom_cmap)
    plt.colorbar()  # Add a colorbar to the right
    plt.title("Continuous Color Map Color List")
```
